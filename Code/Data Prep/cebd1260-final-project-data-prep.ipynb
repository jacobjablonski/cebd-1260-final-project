# Clean the data
I want to predict the GDP of a country. Cleaning this data is going to be tricky because the indicator codes are all in the same column and so are the corresponding values. I need to make new columns for each feature I want to use in my prediction model.

```python
#select what the indicator codes I need
df1 = df_1[(df_1.IndicatorCode == 'SP.DYN.LE00.IN')]
df2 = df_1[(df_1.IndicatorCode == 'EN.ATM.CO2E.PC')]
df3 = df_1[(df_1.IndicatorCode == 'SP.DYN.TFRT.IN')]
df4 = df_1[(df_1.IndicatorCode == 'TX.VAL.MRCH.CD.WT')]
df5 = df_1[(df_1.IndicatorCode == 'NY.GDP.PCAP.CD')]
```

```python
life_expectancy = df1.drop(columns=['CountryName', 'CountryCode', 'IndicatorName', 'IndicatorCode', 'Year'])
life_expectancy.columns = ['life_expectancy']
life_expectancy.reset_index(drop=True)
```

```python
co2_per_capita = df2.drop(columns=['CountryName', 'CountryCode', 'IndicatorName', 'IndicatorCode', 'Year'])
co2_per_capita.columns = ['co2_per_capita']
co2_per_capita.reset_index(drop=True)
```

```python
fertility_rate = df3.drop(columns=['CountryName', 'CountryCode', 'IndicatorName', 'IndicatorCode', 'Year'])
fertility_rate.columns = ['fertility_rate']
fertility_rate.reset_index(drop=True)
```

```python
merch_export = df4.drop(columns=['CountryName', 'CountryCode', 'IndicatorName', 'IndicatorCode', 'Year'])
merch_export.columns = ['merch_export']
merch_export.reset_index(drop=True)
```

```python
gdp = df5.drop(columns=['CountryName', 'CountryCode', 'IndicatorName', 'IndicatorCode', 'Year'])
gdp.columns = ['gdp']
gdp.reset_index(drop=True)
```

```python
a = life_expectancy.join(co2_per_capita, how='outer')
```

```python
b = fertility_rate.join(merch_export, how='outer')
```

```python
c = a.join(b, how = 'outer')
```

```python
d = c.join(gdp, how = 'outer')
```

```python
df_final = d.fillna(d.mean())
df_final
```

```python
test = df_final.iloc[0:11352]
train = df_final.iloc[11353:56760]
```

```python
#Now we save the processed file
train.to_csv('train.csv')
test.to_csv('test.csv')
```
